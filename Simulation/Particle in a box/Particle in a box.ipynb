```python
%matplotlib inline
import time
import random
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib as mpl
import matplotlib.cm as cm
import os, sys
import seaborn as sns; import pandas as pd
plt.rcParams.update({'font.size': 14})

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
    
def calcvelocity(rrange,sigma,vdiff,A,vmin):
    sigmaterm = 2*sigma**2
    distances = np.linspace(0,rrange,100)
    I = A*np.exp(-distances**2/sigmaterm)
    return distances, vdiff*np.exp(-2*I/A)+vmin

#def calcvelocity2(boundaryradius):
#    rrange = np.linspace(0,boundaryradius,100)
#    intensity2 = Imax*lorentz(rrange,Ichar)#Imax*np.exp(-rrange/Ichar)
#    return rrange, (vmax-vmin)*np.exp(-intensity2/Ic)+vmin

def tothecenter(xpos,ypos,angleh):
    value = math.atan(ypos/xpos)*180/math.pi
    if(xpos < 0): angleh = value+180
    else: angleh = value
    if(angleh < 0): angleh += 360
    if(angleh > 360): angleh -= 360
    return angleh, 2

def reflect(xpos,ypos,dirangle):
    value = math.atan(ypos/xpos)*180/np.pi
    angles = np.zeros(3); checks = []
    if(xpos < 0): angles[0] = value+180
    else: angles[0] = value
    if(angles[0] < 0): angles[0] += 360
    if(angles[0] > 360): angles[0] -= 360
    for i in range(1,3): angles[i] = angles[0] + 360 * pow(-1,i)
    for i in range(3): checks.append(abs(angles[i] - dirangle))
    newangle = dirangle + 2 * (angles[checks.index(min(checks))] - dirangle) + 180
    if (newangle > 360): newangle -= 360       
    return newangle, 1

def reflectrand(xpos,ypos,angleh):
    value = math.atan(ypos/xpos)*180/math.pi
    if(xpos < 0): angleh = value+180
    else: angleh = value
    angleh += np.random.uniform(-45,45)
    if(angleh < 0): angleh += 360
    if(angleh > 360): angleh -= 360
    return angleh, 1, value
    
def linear(x, m, n): return m*x+n

def vectorlength(x): return np.sqrt(pow(x[0],2)+pow(x[1],2))
    
def calcintensity(laser,laserframe,delay):
    if delay >= 0:   #positive delay - past   - calc velocity due to sensed intensity[i-delay]
        if laserframe < delay: laser[laserframe,4] = vfactor*math.exp(-2*laser[0,3]/amplitude)+vmin   #if i < delay
        else: laser[laserframe,4] = vfactor*math.exp(-2*laser[laserframe-delay,3]/amplitude)+vmin
    else:            #negative delay - future
        if laserframe >= fitlength:    #extrapolate intensity value
            popt,pcov = curve_fit(linear, fitwave, laser[laserframe-fitlength:laserframe,3],p0 = (0,0))
            futurevalue = linear(-delay,*popt)
            vvalue = vfactor*math.exp(-futurevalue/amplitude)+vmin
            if vvalue >= 0 and vvalue <=1: laser[laserframe,4] = vvalue
            else:                     # 0 < intensity < 1
                if vvalue > 1: laser[laserframe,4] = 1
                if vvalue < 0: laser[laserframe,4] = 0 
        else:     # i < delay
            if laserframe < 2: laser[laserframe,4] = vfactor*math.exp(-2*laser[laserframe,3]/amplitude)+vmin
            else:
                popt,pcov = curve_fit(linear, fitwave[fitlength-laserframe:fitlength], laser[0:laserframe,3],p0 = (0,0))
                futurevalue = linear(-delay,*popt)
                if vvalue >= 0 and vvalue <= 1: laser[laserframe,4] = vvalue
                else:
                    if vvalue > 1: laser[laserframe,4] = 1
                    if vvalue < 0: laser[laserframe,4] = 0  
    return laser

def gauss(x,sigma): return np.exp(-(x/sigma)**2)
def lorentz(x,sigma): return sigma**4/(x**4+sigma**4)

def calcdrifttheory(v,r,tau,delay,base):
    dv = np.gradient(v,base)
    return tau/2*(1-delay/tau)*v*dv+tau*v**2/r

def reflectx(xpos,ypos,dirangle):
    if dirangle >= 0 and dirangle <= 90 : newangle = 180 - dirangle
    if dirangle > 90 and dirangle <= 180: newangle = 180 - dirangle
    if dirangle > 180 and dirangle < 270: newangle = 540 - dirangle
    if dirangle >= 270 and dirangle <= 360: newangle = 540 - dirangle 
    return newangle, 1

def reflecty(xpos,ypos,dirangle):
    if dirangle >= 0 and dirangle <= 90 : newangle = 360 - dirangle
    if dirangle > 90 and dirangle <= 180: newangle = 360 - dirangle
    if dirangle > 180 and dirangle < 270: newangle = 360 - dirangle
    if dirangle >= 270 and dirangle <= 360: newangle = 360 - dirangle
    return newangle, 1
```

```python
np.random.seed(4)
fitwave = np.linspace(-fitlength+1,1,fitlength,endpoint=False)
vfactor=vmax-vmin
laser = np.zeros((stepsexp+1,5))              #laserposition  0 - x; 1 - y; 2 - time at which the laser is set to the new position
                                              #               3 - Ext Int it measures; 4 - theo speed(Int laser)      
anglenoise = np.zeros(stepsexp) #Create random steps
if inputtaur != 0: anglenoise[:] = np.sqrt(2*dt/inputtaur)*57.3*np.random.normal(0,1,stepsexp)          #Create random angle noise
noisesteps = np.zeros((stepsexp,2)) 
for i in range(2): noisesteps[:,i] = prefactor*1e6*np.random.normal(0,1,stepsexp) #Create random steps
deltas = np.zeros(2)
test = np.zeros((stepsexp,8))
test[0,5] = np.random.uniform(0,360)
test[0,0] = np.random.uniform(-0.35*boundaryradius,0.35*boundaryradius)
test[0,1] = np.random.uniform(-0.35*boundaryradius,0.35*boundaryradius)

startpoint = [0]; endpoint = []; out = 0
start_time = time.time() 
for i in range(1,stepsexp):
    test[i,2] = test[i-1,2] + dt                
    test[i,:2] = test[i-1,:2]                              #Position equals position before
    test[i,5] = test[i-1,5]
    if test[i,0] > boundaryradius or test[i,0] < -boundaryradius:
        if test[i,0] > 1.1 * boundaryradius or test[i,0] < -1.1 * boundaryradius: test[i,5], out = tothecenter(test[i,0],test[i,1],test[i,5])
        else:
            if out == 0: 
                test[i,5], out = reflectx(test[i,0],test[i,1],test[i,5])
                endpoint.append(i)
    if test[i,1] > boundaryradius or test[i,1] < -boundaryradius:
        if test[i,1] > 1.1 * boundaryradius or test[i,1] < -1.1 * boundaryradius: test[i,5], out = tothecenter(test[i,0],test[i,1],test[i,5])
        else:
            if out == 0: 
                test[i,5], out = reflecty(test[i,0],test[i,1],test[i,5])
                endpoint.append(i) 
    if out == 2 and -boundaryradius < test[i,0] < boundaryradius and -boundaryradius < test[i,1] < boundaryradius: 
        startpoint.append(i)
        out = 0
    if out == 1: out = 2
    test[i,6] = out
    if out == 0: test[i,5] += anglenoise[i]
    if test[i,5] > 360: test[i,5] -= 360
    if test[i,5] < 0: test[i,5] += 360
    laser[i,0] = test[i,0]+offset*pxtomum*np.cos(test[i,5]*np.pi/180)       
    laser[i,1] = test[i,1]+offset*pxtomum*np.sin(test[i,5]*np.pi/180)
    if test[i,0] < 0: laser[i,3] = 0.8
    else: laser[i,3] = 0.1       
    #laser[i,3] = amplitude*math.exp(-pow(test[i,0]+boundaryradius,2)/sigmafactor)#Imax*lorentz(test[i,0]+boundaryradius,Ichar)
    laser = calcintensity(laser,i,delay)
    if diffusion == True:
        for j in range(2): test[i,j] += noisesteps[i,j]         #add diffusion of the particle
    for j in range(2): deltas[j] = test[i,j]-laser[i,j]
    test[i,3] = np.sqrt(pow(deltas[0],2)+pow(deltas[1],2))
    test[i,4] = maxvelocity
    for j in range(2): test[i,j] += test[i,4]/test[i,3]*deltas[j]*dt*laser[i,4]
if test[i,7] < boundaryradius: endpoint.append(i-1)   
print("--- %s seconds ---" % round((time.time() - start_time),2))
laser[:,4] /= v[-1]
```
